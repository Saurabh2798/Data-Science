# A study about Pixar’s Movies and its influence on Oscar

```python
import pandas as pd # for DataFrame and operations
import matplotlib.pyplot as plt # for plotting base
import seaborn as sns # for compelling plots
%matplotlib inline
```

#### Here are some of the columns in this dataset, PixarMovies.csv:

- Year Released — the year the movie was released. 
- Movie — the name of the movie. 
- RT Score — the Rotten Tomatoes rating for the movie. 
- IMDB Score — the IMDB rating for the movie. 
- Metacritic SCore — the Metacritic rating for the movie. 
- Opening Weekend — the amount of revenue the movie made on opening weekend (in millions of dollars). 
- Worldwide Gross — the total amount of revenue the movie has made to date. 
- Production Budget — the amount of money spent to produce the film (in millions of dollars). 
- Oscars Won — the number of Oscar awards the movie won. 

```python
pixar_movies_df=pd.read_csv("PixarMovies.csv") # read the csv into DataFrame
pixar_movies_df.shape # return the number or rows and columns in the DataFrame
```

```python
pixar_movies_df # show the DataFrame
```

```python
pixar_movies_df.dtypes 
```

```python
pixar_movies_df.describe() # generate summary statistics
```

```python
# converting Domestic and International columns from object to float
pixar_movies_df['Domestic %']=pixar_movies_df['Domestic %'].str.rstrip('%').astype('float')
pixar_movies_df['International %']=pixar_movies_df['International %'].str.rstrip('%').astype('float')
pixar_movies_df.dtypes
```

```python
# converting IMDB score from a 10 point scale to a 100 point scale.
pixar_movies_df['IMDB Score']=pixar_movies_df['IMDB Score']*10
pixar_movies_df
```

```python
filtered_pixar_movies_df=pixar_movies_df.dropna() # as Inside Out in our dataset has NAN values
filtered_pixar_movies_df
```

```python
# changing indexes
pixar_movies_df.set_index('Movie',inplace=True)
filtered_pixar_movies_df.set_index('Movie',inplace=True)
pixar_movies_df
```

## How do Pixar films fare across each of the major review sites?

```python
critics_review=pixar_movies_df[['RT Score','IMDB Score','Metacritic Score']]
critics_review.plot(figsize=(10,6),grid=False)
plt.show()
```

## How are the average ratings from each review site across all the movies distributed?

```python
critics_review.plot(kind='box',figsize=(9,5))
plt.show()
```

## How has the ratio of where the revenue comes from changed since the first movie? Now that Pixar is more well known internationally, is more revenue being made internationally for newer movies?
- To understand how the ratio has changed over the years, we can generate a stacked bar plot. A stacked bar plot is a type of bar plot where the values for each column are stacked on top of each other to better display the ratios. Stacked bar plots are especially useful when plotting proportional values that add upto 100% since it’s easy to visually compare changes across bars.

```python
revenue_proportions=filtered_pixar_movies_df[['Domestic %','International %']]
revenue_proportions.plot(kind='bar',stacked=True)
plt.show()
```

```python
oscar_data=filtered_pixar_movies_df[['Oscars Nominated','Oscars Won']]
oscar_data_plot=oscar_data.plot(kind='bar',figsize=(10,7)) # creates a bar plot
oscar_data_plot.set_xlabel('Movie')
oscar_data_plot.set_ylabel('Total Nomination/Win')
oscar_data_plot.set_title('Comparision between Oscars Nominations/Oscars Wins')
```
