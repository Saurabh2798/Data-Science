```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("E:/Data Science/kaggle/kaggle-data/digit-recognizer/train.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
features=df.loc[:,df.columns!='label']
labels=df['label']
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,labels,test_size=0.1)
```

```python
X_train = X_train.values.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.values.reshape(X_test.shape[0], 28, 28, 1)
X_train.shape
```

```python
num_classes=10
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
y_train.shape
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(28,28,1)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train,
          batch_size=128,
          epochs=20,
          verbose=1,
          validation_data=(X_test, y_test))
```

```python
test_df=pd.read_csv("E:/Data Science/kaggle/kaggle-data/digit-recognizer/test.csv")
```

```python
preds=model.predict(test_df.values.reshape(test_df.shape[0],28,28,1))
preds=np.argmax(preds,axis=1)
preds
```

```python
pd.DataFrame(preds,columns=['Label']).to_csv("output.csv")
```
